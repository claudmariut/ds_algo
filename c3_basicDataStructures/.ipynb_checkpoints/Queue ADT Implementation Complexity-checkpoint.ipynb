## Queue Linear Data Structure
##### Operation Performance

The two basic operations of the Queue are .dequeue() and .enqueue(). There are two different ways to implement the Queue data structure using the list built in data structure in python. 

First, we can think consider the rear of the Queue as the beginning of the list, and the front as the end of the list. In this case: .enqueue() operation will be implemented using .insert(0, item), and .dequeue() operation will be implemented using .pop(). Understanding the list functions time complexity we can see that for this Queue implementation .dequeue() behaves with O(1) constant time, and .enqueue() with O(n) linear time.

- We will call this first implementation q1. "q1" as being: q1 = Queue()

Second, we can consider the rear of the Queue as the end of the list, and the front as the beginning of the list. In this implementation, for .dequeue() operation, we will use .pop(0), which is O(n) linear time, and for .enqueue() operation, we will use .append(item), which is O(1) constant time.

- We will call this second implementation q2. "q2" as being: q2 = Queue2()

Note that Queue and Queue2 are different classes, both represent a Queue but with different implementation. q1 and q2 are object of this classes respectively.

### Queue Implementation O(1) Avg for Dequeue() and Enqueue() operations.

Using a linked list, implementing a new attribute for the linked list class called
self.end that always points to the end of the linked list. Usually a single linked list would have a pointer to the head node. (self.head) definded in the constructor of the class. Each node would be linked at the end of the last one. Usually, this operation takes linear amount of time since the linked list is an unordered data structure. The self.end attribute is always pointing to the last node in the linked list. 

Each time, we want to use the enqueue() function, the new item would be linked at the end of the linked list accesing the last node directly in constant time. When we use dequeue() function, we return the value of the head of the linked list in constant time, and set the new head to the next element.
